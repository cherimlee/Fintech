### 이미지 읽기, 쓰기 및 표시하기 / 그레이 스케일 변환

```python
import numpy as np
import cv2

def handle_image():
    imgfile = './Selected/000.jpeg'
             #이미지를 읽기 위한 객체 리턴 
                    #파일경로, 컬러이미지로 불러옴    
    img = cv2.imread(imgfile, cv2.IMREAD_GRAYSCALE)
                              # 크기를 조절할 수 있는 윈도우 생성 
    cv2.namedWindow("image", cv2.WINDOW_NORMAL)   # WINDOW_AUTOSIZE : 원본 이미지 크기로 보정해서 윈도우 생성
    # 이미지 출력 , 타이틀, 이미지 객체  
    cv2.imshow('image', img)
    # 대기하는 시간 (밀리세컨드)
    k = cv2.waitKey(0)
    
    # wait for ESC key to exit
    if k == 27 :
        cv2.destroyAllWindows()
        cv2.waitKey(1)
        
    # wait for 's' key to save and exit
    elif k == ord('s'):
        cv2.imwrite("grayImage.png", img)    # grayImage라는 이름으로 저장
        cv2.destroyAllWindows()
        cv2.waitKey(1)
    
if __name__ == '__main__':
    handle_image()
    
```

### 도형 외곽 추출하기 

```python
import numpy as np
import cv2
from matplotlib import pyplot as plt

def contour():
    imgfile = './Selected/000.jpeg'
    img = cv2.imread(imgfile)
    
    # 흑백이미지를 따로 생성 
    imgray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)   # 색공간을 BGR에서 gray로 변환
    
    edge = cv2.Canny(imgray, 100, 200)   # edge detection에 대한 threshold
    contours, hierarchy = cv2.findContours(edge, cv2.RETR_TREE, cv2.CHAIN_APPROX_SIMPLE)
    
    cv2.imshow('edge', edge)
    cv2.drawContours(img, contours, -1, (0, 255, 0), 1)
    cv2.imshow('Contour', img)
    
    cv2.waitKey(0)
    cv2.destropAllWindows()
    cv2.waitKey(1)
    
if __name__ == '__main__':
    contour()
```

### 도형이 구겨져 있거나 틀어져 있을 경우 ?
### 도형을 근사해서 overfit을 방지합니다. 

```python
import numpy as np
import cv2
from matplotlib import pyplot as plt

def contour_approx():
    imgfile = './Selected/000.jpeg'
    img = cv2.imread(imgfile)
    img2 = img.copy()
    imgray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    
    edge = cv2.Canny(imgray, 100, 200)
    contours, hierarchy = cv2.findContours(edge, cv2.RETR_TREE, cv2.CHAIN_APPROX_SIMPLE)
    
    cnt = contours[0]
    cv2.drawContours(img, [cnt], 0, (0, 255, 0), 3)
    
    # 컨투어의 둘레의 길이 계산 (둘레 길이의 10%를 입실론으로 활용)
    # 입실론 : 근사 정확도 
    epsilon = 0.1 * cv2.arcLength(cnt, True)
    
    # 오차를 최대한으로 꼭지점을 줄여나감 (입실론 값이 작을수록 원래 컨투어와 비슷)
    approx = cv2.approxPolyDP(cnt, epsilon, True)
    
    cv2.drawContours (img2, [approx], 0, (0, 255, 0), 3)
    
    cv2.imshow('Contour', img)
    cv2.imshow('Approx', img2)
    
    cv2.waitKey(0)
    cv2.destropAllWindows()
    cv2.waitKey(1)
    
if __name__ == '__main__':
    contour_approx()
```

### 투영 변환 
* 외곽으로 그려진 영역을 반듯하게 변환하기 (특정 픽셀을 어디에서 어디로 옮길지 결정)
* 4개의 점 필요

```python
import numpy as np
import cv2

def warpAffine():
    img = cv2.imread("./Selected/000.jpeg")
    
    # 세 개의 좌표 
    pts1 = np.float32([[50, 50], [200, 50], [20, 200]])
    pts2 = np.float32([[70, 100], [220, 50], [150, 250]])
    
    # pts1 -> pts2 로 옮기는 메트릭스 반환  
    M = cv2.getAffineTransform(pts1, pts2)
    
    #                               이미지 크기 
    result = cv2.warpAffine(img, M, (350, 300))
    
    cv2.imshow("original", img)
    cv2.imshow("Affine Rransform", result)
    
    cv2.waitKey(0)
    cv2.destroyAllWindows()
    cv2.waitKey(1)
    
if __name__ == '__main__':
    warpAffine()
```

### 투영변환 : Threshold를 조정하여 조명의 영향을 최소화 
* 이미지를 잘게 쪼개서 Threshold 구하기 : adaptive threshold
* 노이즈가 많을 수록 이진화를 하기 전에 블러효과를 주지 않으면 결과가 안좋을 수 있음 : 블러 효과로 완만하게 변환 

```python
def adaptive_threshold():
    imgfile = "./Selected/000.jpeg"
    img = cv2.imread(imgfile, cv2.IMREAD_GRAYSCALE)
    
    r = 600.0 / img.shape[0]
    dim = (int(img.shape[1] * r), 600)
    img = cv2.resize(img, dim, interpolation = cv2.INTER_AREA)
    
    # 이미지에 블러 효과 주는 법 : 주변 픽셀의 평균 값을 대입 
    blur = cv2.GaussianBlur(img, (5, 5), 0)   # (9,9) -> 블러 효과 9만큼 줬을 때 
    result_without_blur = cv2.adaptiveThreshold(img, 255, cv2.ADAPTIVE_THRESH_MEAN_C, cv2.THRESH_BINARY, 21, 10)
    result_with_blur = cv2.adaptiveThreshold(img, 255, cv2.ADAPTIVE_THRESH_MEAN_C, cv2.THRESH_BINARY, 21, 10)

    cv2.imshow("Without Blur", result_without_blur)
    cv2.imshow("With Blur", result_with_blur)
    
    cv2.waitKey(0)
    cv2.destroyAllWindows()
    cv2.waitKey(1)
    
if __name__ == '__main__':
    adaptive_threshold()
```

```python
image = cv2.imread("./Selected/000.jpeg")
orig = image.copy()          # 오리지널 이미지는 복사해서 저장 
r = 800.0 / image.shape[0]
dim = (int(image.shape[1] * r), 800)
image = cv2.resize(image, dim, interpolation = cv2.INTER_AREA) # 이미지 리사이즈 

gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)  # 그레이 컬러로 변환
gray = cv2.GaussianBlur(gray, (3, 3), 0)        # 블러 효과 주기
edged = cv2.Canny(gray, 75, 200)                # 엣지 검출 

print ("STEP 1: Edge Detection")
cv2.imshow("Image", image)
cv2.imshow("Edged", edged)

cv2.waitKey(0)
cv2.destroyAllWindows()
cv2.waitKey(1)
```
