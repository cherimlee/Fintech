```python
import pandas as pd
import nltk
from nltk.stem.porter import PorterStemmer
from nltk.tokenize import RegexpTokenizer
from nltk.corpus import stopwords
```

```python
f = open('data_files/trump_speech.txt', 'r', encoding='utf-8')
lines = f.read()
f.close()
```

```python
print(lines)
```

```python
tokenizer = RegexpTokenizer('[\w]+')
```

```python
# 대소문자 통일 - 영문
words = lines.lower()
```

```python
# 토큰 단위로 구분 ( 한글 : 형태소 분리)
tokens = tokenizer.tokenize(words)
tokens
```

```python
nltk.download('stopwords')
```

```python
# 불용어 : 전체 문맥에서 의미를 갖지 않는 토큰( 형태소)

stop_words = stopwords.words('english')
stopped_tokens = [i for i in list(tokens) if not i in stop_words]
```

```python
stopped_tokens1 = [i for i in stopped_tokens if len(i) > 1]  # 영문
```

```python
word_count = pd.Series(stopped_tokens1).value_counts()
word_count.head(10)
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(20,5))
word_count[:20].plot(kind='bar')
```
