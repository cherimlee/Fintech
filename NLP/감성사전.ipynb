```python
def emotion_score(series):
    
    emot_score=0
    
    for j in range(len(series)):
        
        if series.iloc[j] in ('positive','anticipation','trust','joy','surprise'):
            emot_score += series.iloc[j]
        else :
            emot_score -= series.iloc[j]
        return emot_score
```
