```python
movie_actor_director = movie[['actor_1_name', 'actor_2_name', 'actor_3_name', 'director_name']]
actor = ['actor_1_name', 'actor_2_name', 'actor_3_name']
direc = ['director_name']
sort_set = (direc + actor)
movie[sort_set]
# movie_actor_director

# for 문을 사용하는 방법
li = []
for i in movie_actor_director.columns:
    if i == 'director_name':
        li.append(i)
    else:
        li.append(i)
        
movie[li]
```

```python
college = pd.read_csv('./data/college.csv', index_col = 'INSTNM')
college.head()
```

drop()

```python
coll = pd.read_csv('./data/college.csv', index_col='INSTNM')
coll = coll.filter(like='UGDS_')
coll
```

```python
# 모든 Nan값을 drop하면 다양한 계산이 가능
# any: 모든 행의 컬럼값 중 하나라도 NULL의 값을 가지면 해당 행을 제거
# all: 모든 칼럼 값이 NULL이거나 NaN인 경우에만 해당 행을 제거


coll = coll.dropna(how='all')
coll = coll.dropna(how='any')
coll.isnull().sum()
```
