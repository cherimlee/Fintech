```python
# 리튬

df_li = pd.read_excel('./Data/리튬 가격2022-03-07.xlsx')
df_li.head(5)
```

```python
df_li.drop(['등락가', '등락비율', '재고량'], axis = 1, inplace=True)
df_li
```

```python
df_li.columns = 'date', '리튬가격'
df_li
```

```python
# 코발트

df_co = pd.read_excel('./Data/코발트 가격2022-03-07.xlsx')
df_co.head(5)
```

```python
df_co.drop(['등락가', '등락비율', '재고량'], axis = 1, inplace = True)
df_co
```

```python
df_co.columns = 'date', '코발트가격'
df_co
```

```python
# 희토류

df_ra = pd.read_excel('./Data/희토류 가격2022-03-07 (1).xlsx')
df_ra.head(5)
```

```python
df_ra.drop(['등락가', '등락비율', '재고량'], axis = 1, inplace = True)
df_ra
```

```python
df_ra.columns = 'date', '희토류가격'
df_ra
```
