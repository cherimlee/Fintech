```python
df_mine = df[['date', 'tesla', 'tesla_거래량', 'tesla_volatility','EVE종가', 'EVE거래량', 'EVE변동률','BYD종가',\
'BYD거래량', 'BYD변동률','LIT조정종가', 'LIT거래량', 'SGL종가', 'SGL변동률','li_price',\
'nasdaq', 'nasdaq_거래량', 'nasdaq_volatility',]]
```

```python
y = df_mine['tesla']
x = df_mine.drop(['tesla', 'date'], axis = 1)
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, train_size=0.2, random_state=111)
```

```python
x_train.info()
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
log = LogisticRegression()
```

```python
log.fit(x_train, y_train)
```

```python
preds = log.predict(x_test)
```

```python
preds
```

```python
print('train acc :', log.score(x_train, y_train)) 
print('test acc :', log.score(x_test, y_test))
```

```python
from sklearn.metrics import classification_report, confusion_matrix
```

```python
confusion_matrix(y_test, preds)
```

```python
print(classification_report(y_test, preds))
```
