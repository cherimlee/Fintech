```python
def finance_data_preprocessing(path):
    
    raw_data = pd.read_excel(path)
    
    big_col = list(raw_data.columns)
    for num, temp in enumerate(big_col):
        if '.' in temp:
            big_col[num] = temp.split('.')[0]
            
    small_col = list(raw_data.loc[0])
    small_col[0] = big_col[0]
    small_col[1] = big_col[1]
    
    raw_data.columns = [big_col, small_col]
    raw_data = raw_data.drop(0)
    
    raw_data.index = raw_data['Symbol']['Symbol']
    del raw_data['Symbol']
    
    return raw_data
```

```python
def small_size(path, date):
    data_df = finance_data_preprocessing(path)

    date = date
    fin_df = data_df[[('시총', date), ('PER',  date), ('PBR', date), ('PSR', date), ('Symbol Name', 'Symbol Name')]]
    fin_df['시총'][date] = pd.to_numeric(fin_df['시총'][date])
    
    market_cap_quantile_series = fin_df['시총'][date].quantile(0.2)
    filtered_df = fin_df[fin_df['시총'][date] <= market_cap_quantile_series]
    
    filtered_df = filtered_df[(filtered_df['PBR'][date] >= 0.2) & (filtered_df['PER'][date] <= 0.8)]
    
    return filtered_df['PBR'][date].sort_values()[:10]
```

```python
def benjamin_grahm(path, date):
    data_df = finance_data_preprocessing(path)

    fin_df = data_df[[('시총', date), ('ROA',  date), ('PBR', date), ('부채비율', date), ('Symbol Name', 'Symbol Name')]]
    fin_df['시총'][date] = pd.to_numeric(fin_df['시총'][date])
    
    filtered_df = fin_df[fin_df['ROA'][date] >= 0.05]
    filtered_df = filtered_df[filtered_df['부채비율'][date] <= 100]
    filtered_df = filtered_df[filtered_df['PBR'][date] >= 0.2]
    filtered_df = filtered_df['PBR'][date].sort_values()[:10]
    
    return filtered_df
```

```python
def pfo_backtest(st_list, init_money, backtest_price, st_date, ed_date):

    selected_list = st_list

    each_money = init_money/len(selected_list)
    backtest_price = backtest_price[st_date: ed_date]

    pf_value = 0
    for code in selected_list:
        temp_num = int(each_money/int(backtest_price.iloc[0][code]))  # 매입종목수
        pf_value = pf_value + pd.to_numeric(backtest_price[code]) * temp_num  # pfo의 현재가치

    cash = init_money - pf_value[0]
    cash_pf = []

    for i in range(0, len(pf_value)):
        cash_pf.append(cash * (1 + riskfree_rate/12)**i)

    backtest_df = pd.DataFrame({'주식포트': pf_value})
    backtest_df['현금포트'] = cash_pf
    backtest_df['종합포트'] = backtest_df['주식포트'] + backtest_df['현금포트']
    
    backtest_df['종합포트'].plot()

    return backtest_df
```

```python
st_list = ['A000020', 'A000030', 'A000040', 'A000050', 'A000060']
init_money = 100000000
backtest_price = price_data_preprocs('수정주가.csv')
st_date = '2015-01'
ed_date = '2019-12'

pfo_backtest(st_list, init_money, backtest_price, st_date, ed_date)
```
