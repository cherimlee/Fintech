```python
# 최대값, 최소값 찾기
import numpy as np

num_list = []

for _ in range(20):
    num = np.random.randint(1,50)
    num_list.append(num)
    
num_list
```

```python
# 최댓값과 최솟값 구하기

max_num = 0
min_num = 0

for n in num_list :
    if max_num > n :
        max_num = n
    if min_num < n:
        min_num = n
        
print(max_num, min_num)
```

```python
import os
```

```python
files = os.listdir('./names')
```

```python
year_sex_pop = {}

for file in files :
    with open("./names/" + file, 'r') as f:
        lines = f.readlines()
        
        f_pop = 0
        m_pop = 0
        
        for line in lines :
            sex = line.split(',')[1]
            pop = line.split(",")[2]
            
            if sex == 'F' :
                f_pop += int(pop)
            else :
                m_pop += int(pop)
                
        year_sex_pop[file + 'f'] = f_pop
        year_sex_pop[file + 'm'] = m_pop
```
