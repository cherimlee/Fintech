```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
cols = ['pregnant','piasma','pressure','thinkness','insuline','bmi','pedigree','age','class']
df = pd.read_csv('./deep_source_files/pima-indians-diabetes.csv', header = None, names = cols)
df
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
import tensorflow as tf
```

```python
dataset = np.loadtxt('./deep_source_files/pima-indians-diabetes.csv', delimiter = ',')
dataset
```

```python
y = dataset[:,8]
x = dataset[:,0:8]
```

```python
model = Sequential()
model.add(Dense(12, input_dim=8, activation='relu'))  # input layer
model.add(Dense(8, activation='relu'))                # hidden layer
model.add(Dense(1))                                   # output layer
# binary classification 에서는 아웃풋 레이어에 1,2 사용 가능 

# model 구조 
model.summary()
```

```python
 model.compile(loss='binary_crossentropy', metrics=['accuracy'], optimizer = 'adam')
                                                                # optimizer : gradientdescent의 방식 
```

```python
model.fit(x,y, epochs=100, batch_size=20)

# epochs : 전체 데이터를 몇 번 학습 시킬지 
# batch_size : 한 번 학습 시킬 때 몇 개의 데이터를 학습시킬 것인가 
# batch size가 크면 업데이트 시 메모리와 시간이 많이 소요 / 업데이트는 안정적
# batch size가 작으면 업데이트 시 메모리와 시간이 작게 소요 / 업데이트는 불안정적
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
x_scaled = scaler.fit_transform(df.iloc[:,:8])
```
