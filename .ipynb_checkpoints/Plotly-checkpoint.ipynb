```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
import plotly.express as px
```

```python
df = px.data.stocks()
df
```

```python
df.info()
```

```python
px.line(df, x='date', y='GOOG')
```

```python
df = df.set_index("date")
```

```python
# 가격을 2018-01-01 기준을 1로 잡고 정규화 
df
```

```python
df = df - 1
```

```python
# 변화율
px.bar(df, x=df.index, y='GOOG')
```

```python
df.columns.name = 'company'
df.columns
```

```python
px.area(df, facet_col='company')
```

```python
px.area(df, facet_col='company', facet_col_wrap=2)
```

```python
px.line(df, x=df.index, y=df.columns)
```

```python
px.line(df, x=df.index, y='AAPL')
```

```python
# rangeslider 기간을 설정할 수 있음
fig = px.line(df, x=df.index, y='AAPL')
fig.update_xaxes(rangeslider_visible=True)
```

```python
import plotly.graph_objects as go
```

```python
df = pd.read_csv("https://raw.githubusercontent.com/plotly/datasets/master/finance-charts-apple.csv")
df
```

```python
fig = go.Figure(data=[go.Candlestick(x=df['Date'],
                              open = df['AAPL.Open'],  # 시가
                              high = df['AAPL.High'],  # 고가
                              low = df['AAPL.Low'],   # 저가
                              close = df['AAPL.Close'])])  # 종가
fig.show()
```
